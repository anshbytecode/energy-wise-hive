```python
# Sustainable Energy & Efficiency Project
# Author: Anshul Bhilare

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

years = np.arange(2010, 2025)
data = {
    "Year": years,
    "Solar": np.linspace(20, 150, len(years)) + np.random.randint(-10, 10, len(years)),
    "Wind": np.linspace(30, 180, len(years)) + np.random.randint(-15, 15, len(years)),
    "Hydro": np.linspace(100, 120, len(years)) + np.random.randint(-5, 5, len(years)),
    "Fossil": np.linspace(300, 200, len(years)) + np.random.randint(-20, 20, len(years))
}

df = pd.DataFrame(data)
df["Total"] = df[["Solar", "Wind", "Hydro", "Fossil"]].sum(axis=1)
df.head()
```

```python

latest = df.iloc[-1]  # latest year
sources = ["Solar", "Wind", "Hydro", "Fossil"]

plt.figure(figsize=(6,6))
plt.pie(latest[sources], labels=sources, autopct='%1.1f%%', startangle=90)
plt.title("Energy Mix in " + str(latest["Year"]))
plt.show()
```

```python

plt.figure(figsize=(10,5))
plt.plot(df["Year"], df["Solar"], label="Solar")
plt.plot(df["Year"], df["Wind"], label="Wind")
plt.plot(df["Year"], df["Hydro"], label="Hydro")
plt.plot(df["Year"], df["Fossil"], label="Fossil Fuels")

plt.xlabel("Year")
plt.ylabel("Energy (TWh)")
plt.title("Energy Production Over Time")
plt.legend()
plt.show()
```

```python

df["CO2_Emissions"] = df["Fossil"] * 0.8  

plt.figure(figsize=(10,5))
plt.bar(df["Year"], df["CO2_Emissions"], color="red")
plt.xlabel("Year")
plt.ylabel("CO2 Emissions (Million tons)")
plt.title("Carbon Emissions Trend")
plt.show()
```

```python

df["Renewables"] = df["Solar"] + df["Wind"] + df["Hydro"]
df["Efficient_Renewables"] = df["Renewables"].where(df["Year"]<2020,
                                                    df["Renewables"]*1.1)

plt.figure(figsize=(10,5))
plt.plot(df["Year"], df["Renewables"], label="Before Efficiency")
plt.plot(df["Year"], df["Efficient_Renewables"], label="After Efficiency", linestyle="--")
plt.xlabel("Year")
plt.ylabel("Renewable Energy (TWh)")
plt.title("Impact of Efficiency Improvements")
plt.legend()
plt.show()
```

```python

print("✅ Renewable energy is steadily increasing while fossil fuels are decreasing.")
print("✅ Efficiency improvements lead to higher renewable output with the same resources.")
print("✅ Carbon emissions decrease as reliance on renewables grows.")
```
